# Math operations

operations in columns

```python
#selct a column and calculate total sum
df_exams['math score'].sum()
```

```python
#the number of rows count, std, max and min
df_exams['math score'].count()
df_exams['math score'].mean()
df_exams['math score'].max()
df_exams['math score'].min()
df_exams['math score'].std()
```

```python
#easier calculation with .describe()
df_exams.describe()
```

**Work with rows**

```python
#calculating the sum in a row
df_exams['math score'] +df_exams['reading score']+df_exams['writing score']
```

```python
#average in a row
df_exams['average']=(df_exams['math score'] +df_exams['reading score']+df_exams['writing score'])/3
df_exams.round(2)
```

# Sort the dataframe

```python
#sort by one column
df_exams.sort_values(by='math score')
```

```python
#sort in descending order by one column
df_exams.sort_values(by='math score',ascending=False)
```

```python
#sort descending by multiples columns
df_exams.sort_values(by=['math score','reading score'],ascending=False,inplace=True)
df_exams
```

```python
#sort descending with a key function
df_exams.sort_values('race/ethnicity',ascending=True,key=lambda col: col.str.lower())
```
